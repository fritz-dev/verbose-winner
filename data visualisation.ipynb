```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('table1and2.csv')
```

```python
auk = data[data.Region == 'AUK']
auk = auk[auk.IncomeCode == 10]
```

```python
wel = data[data.Region == 'WEL']
wel = wel[wel.IncomeCode == 10]
```

```python
#grouping by Region and Year for AUK and WEL
group_auk = auk.groupby(['Region','Year']).agg({'Mean':['mean']})
group_auk.columns = ['Mean']
group_auk = group_auk.reset_index()
df_auk = pd.DataFrame(group_auk)

group_wel = wel.groupby(['Region','Year']).agg({'Mean':['mean']})
group_wel.columns = ['Mean']
group_wel = group_wel.reset_index()
df_wel = pd.DataFrame(group_wel)

```

```python
#plt.plot(df_auk.Year,df_auk.Mean)
plt.plot(df_auk.Year,df_auk.Mean/df_auk.Mean.iloc[0])
#plt.plot(df_wel.Year,df_wel.Mean)
plt.plot(df_wel.Year,df_wel.Mean/df_wel.Mean.iloc[0])
plt.title('Average Household Salary Income Auckland vs Wellington')
plt.legend(['Auckland', 'Wellington'])
plt.xlabel('Year')
plt.ylabel('Salary Income')
plt.show()
```
